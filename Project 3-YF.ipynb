```python
#add a new column to show "Date" column as Quarter
close['Quarter'] = close['Date'].dt.to_period('Q')
close
```

```python
#rename the Quater colummn as current Quarter
close.rename(columns = {'Quarter' : 'Cur Quarter'}, inplace = True)
close
```

```python
#this formula use to differentiation the previous quarter vs current quater, previous quater is a "date value" to calcuate PE ratio for current trading price, which use close price
def previous_quarter_date(val):
    if val.month < 4:
        return datetime.date(val.year - 1, 12, 31)
    elif val.month < 7:
        return datetime.date(val.year, 3, 31)
    elif val.month < 10:
        return datetime.date(val.year, 6, 30)
    return datetime.date(val.year, 9, 30)
```

```python
#assign value to previous quarter
import datetime
close['pre Qdate'] = close['Date'].apply(previous_quarter_date)
close
```

```python
#adding a column as of FSQuarter for merge purpose,format to same format of Current Quarter
close['pre Qdate'] = pd.to_datetime(close['pre Qdate'])
close['FSQuarter'] = close['pre Qdate'].dt.to_period('Q')
close['year'] = close['Date'].dt.year
close
```

```python
#drop the 'pre Qdate'column for data cleaning
column_to_drop = 'pre Qdate'
close.drop(columns=[column_to_drop], inplace=True)
close
```

```python
#import AMZN earning report Data set from Fidelity account
amznEPS = pd.read_csv("amzn.csv")
amznEPS
```

```python
print(amznEPS.columns)
```

```python
#add column as of FSQuarter for merge purpose
quarter_info = amznEPS['Quarter'].str.extract(r'Q(\d) (\d{4})')
amznEPS['FSQuarter'] = quarter_info[1] + 'Q' + quarter_info[0]
```

```python
amznEPS
```

```python
#check datatype for merge purpose
close["FSQuarter"].info()
```

```python
amznEPS["FSQuarter"].info()
```

```python
#change the merge column to the same datetype 
amznEPS['FSQuarter'] = pd.PeriodIndex(amznEPS['FSQuarter'], freq='Q-DEC')
amznEPS["FSQuarter"].info()
```

```python
#drop the n/a rows and adding year column 
amznEPS = amznEPS.dropna()
amznEPS['year'] = amznEPS['Quarter'].str.extract('(\d{4})').astype(int)
amznEPS
```

```python
#change the value to number for later calculation 
#yeps value for Earning per share(calendar year, chart 1)
#4Q Sum, value for Earning per share(last four Quarterly report, PE ration, chart2)
amznEPS['Reported EPS ($)'] = pd.to_numeric(amznEPS['Reported EPS ($)'], errors='coerce')
amznEPS['amznyeps'] = amznEPS.groupby('year')['Reported EPS ($)'].transform('sum')
amznEPS
```

```python
#to get value for all 4 previous quarterly eps as rolling annual eps
data = amznEPS["Reported EPS ($)"]
amznEPSdata = pd.DataFrame(data)
amznEPSdata['Reported EPS ($)'] = pd.to_numeric(amznEPSdata['Reported EPS ($)'], errors='coerce')
window_size = 4
rolling_sum = amznEPSdata.rolling(window=window_size).sum()
rolling_sum_shifted = rolling_sum.shift(-window_size + 1)
```

```python
amznEPS['amzn4Qsum'] = rolling_sum_shifted
amznEPS
```

```python
merged_amznEPS = pd.merge(close, amznEPS, on='FSQuarter', how='inner')
merged_amznEPS
```

```python
#Merege the EPS to Master dataset
close["amzn4Qsum"] = merged_amznEPS['amzn4Qsum']
close["amznPEratio"] = close["AMZN"] / close["amzn4Qsum"]
```

```python
close
```

```python
#To get the value PE Ration as financial mertic
close['amznyeps'] = merged_amznEPS['amznyeps']
close
```
